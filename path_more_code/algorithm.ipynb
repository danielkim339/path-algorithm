```python
import pygame


import game_graphs


class PathSearch:
    
    global running, g
    
    def __init__(self, start, goal, alg):
        self.start = start
        self.goal = goal
        self.alg = alg
        self.frontier = Queue() if alg=='bfs' else PriorityQueue()
        self.came_from = {}
        self.came_from[start] = None 
        self.cost_so_far = {}
        self.cost_so_far[start] = 0
        self.finish = False    
        self.search_loop(alg)
        
    def search_loop(self, alg):
        g.update_grid()
        while not self.frontier.empty() and running:                  
            current = self.frontier.get()
            if current == self.goal:
                self.finish = True                                                         
                return self.finish 
            
            if alg == 'bfs':
                current_frontier, current_visited = breadth_first_search()
            elif alg == 'greedy':
                current_frontier, current_visited = greedy()
            elif alg == 'astar':
                current_frontier, current_visited = a_star()
            elif alg == 'dij':
                current_frontier, current_visited = dijkstras()
                
            g.frontier = current_frontier          
            g.visited = current_visited            
            g.add_frontier()    
            
            if self.finish == False:
                move = False
                while move == False:
                    # pygame.event.pump()
                    pygame.event.get()
                    key = pygame.key.get_pressed()
                    if key[pygame.K_SPACE]:
                        self.finish = True
                        break
                    elif key[pygame.K_RIGHT]:
                        move = True
                        g.update_grid() 
                    elif key[pygame.K_ESCAPE]:
                        running = False
                        break            

    def breadth_first_search(self):
        for n in g.neighbors(current):
            if n not in came_from:    
                frontier.put(n)
                came_from[n] = current
        current_frontier = list(frontier.elements)                                       
        current_visited = list(set(key for key in came_from))    
        return current_frontier, current_visited
    
    def greedy(self):
        for n in g.neighbors(current):
            if n not in came_from:               
                priority = heuristic(goal, n)       
                frontier.put(n, priority)
                came_from[n] = current
        current_frontier = [item[1] for item in frontier.elements]               #####        
        current_visited = list(set(key for key in came_from)) 
        return current_frontier, current_visited

    
    def dijkstras(self):
        for n in g.neighbors(current):
            new_cost = cost_so_far[current] + g.cost(current, n)   
            if n not in cost_so_far or new_cost < cost_so_far[n]:
                cost_so_far[n] = new_cost
                priority = new_cost
                frontier.put(n, priority)
                came_from[n] = current
        
        current_frontier = [item[1] for item in frontier.elements]         
        current_visited = list(set(key for key in came_from))  
        return current_frontier, current_visited
                
    def a_star(self):
        for n in g.neighbors(current):
            new_cost = cost_so_far[current] + g.cost(current, n)
            if n not in cost_so_far or new_cost < cost_so_far[n]:
                cost_so_far[n] = new_cost
                priority = new_cost + heuristic(goal, n)
                frontier.put(n, priority)
                came_from[n] = current
            
        current_frontier = [item[1] for item in frontier.elements]        
        current_visited = list(set(key for key in came_from)) 
        return current_frontier, current_visited
    
    def end_search(self):
        if frontier.empty() and current != goal:   ### SINCE frontier is DICTIONARY, CHECK DEAD END IN BETTER WAY
            pygame.display.set_caption('Dead end. Try again.')
            return None, None 
        else:
            return came_from, cost_so_far
    
        
```
